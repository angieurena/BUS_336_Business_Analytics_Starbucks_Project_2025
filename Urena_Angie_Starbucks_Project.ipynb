```python
import pandas as pd

df = pd.read_csv('BUS336_Signature_Term_Project_Starbucks_Beverage_Menu.csv')
print("Dataset shape:", df.shape)
df.head()
```

```python
print("NaN values per column:")
df.isnull().sum()
```

```python
df.head(15)
```

```python
df1 = df.dropna(how='all')

print("After removing blank rows:")
df1.isnull().sum()
```

```python
df1.info()
```

```python
df1.iloc[1,13] = 0
```

```python
df1['Sugar_g'] = df1['Sugar_g'].astype(float)
```

```python
print("Sugar_g data type:", df1.Sugar_g.dtypes)
```

```python
df1.info()
```

```python
df2 = df1.dropna()
print("df2 created successfully!")
print(f"df2 shape: {df2.shape}")
```

```python
df2.info()
```

```python
df3 = df2.rename(columns={'Caffiene_milligrams': 'Caffiene_mg'})
print("df3 created successfully!")
```

```python
df3.head()
```

```python
print("Continuous Variables Summary:")
df3.describe().round(2)
```

```python
print("Categorical Variables Summary:")
df3.describe(include='object')
```

```python
print("Calories statistics:")
print(f"Min: {df3['Calories'].min()}, Max: {df3['Calories'].max()}, Mean: {df3['Calories'].mean():.1f}")

print("\nSugar statistics:")
print(f"Min: {df3['Sugar_g'].min()}, Max: {df3['Sugar_g'].max()}, Mean: {df3['Sugar_g'].mean():.1f}")

print("\nCaffeine statistics:")
print(f"Min: {df3['Caffiene_mg'].min()}, Max: {df3['Caffiene_mg'].max()}, Mean: {df3['Caffiene_mg'].mean():.1f}")
```

```python
# Analysis and Interpretation Comments:
# The Starbucks beverage menu contains 1203 total items after cleaning
# Calories range from 0 to 640 with a mean of 230.8 calories per beverage
# The most common beverage product appears to be White Hot Chocolate with 50 occurrences
# Sugar content varies significantly from 0g to 89g with an average of 35.0g per drink
# Caffeine levels range from 0mg to 475mg with most beverages averaging 82.0mg
# Size options include 7 varieties with Grande being the most popular choice (345 occurrences)
# Milk preferences show 6 options available with Nonfat being most common (240 selections)
# Most beverages are served without whip cream (910 out of 1203 items)
# The data suggests Starbucks offers high-calorie options with some drinks reaching 640 calories
# Coffee beverages tend to have moderate caffeine levels compared to energy drinks in the market
```
